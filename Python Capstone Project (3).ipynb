```python
numerical_columns = df.select_dtypes(include=['int64', 'float64'])
categorical_columns = df.select_dtypes(include=['object'])

from tabulate import tabulate
table = [numerical_columns, categorical_columns]
print(tabulate({"Numerical":numerical_columns,
                "Categorical": categorical_columns}, headers = ["numerical_columns", "categorical_columns"]))
```

# Separating Categorical and Numerical columns

```python
df.isnull().sum()
```

```python
# Fill missing values in numerical columns with the median
for column in numerical_columns:
    mean = df[column].mean()  
    df[column].fillna(mean, inplace=True) 
```

```python
for column in categorical_columns:
    mode_value = df[column].mode().iloc[0]  
    df[column].fillna(mode_value, inplace=True)  
```

```python
df.isnull().sum()
```

# Identifying and displaying production companies along with their counts

```python
import json
# Convert the string representation of production companies into a list of dictionaries
df['production_companies'] = df['production_companies'].apply(lambda x: json.loads(x))
```

```python
# # Initialize an empty dictionary to store production company names and their counts
production_counts = {}

# Iterate over each row in the DataFrame
for row in df.itertuples(index=False):
    # Iterate over each production company in the row
    for company in row.production_companies:
        # Extract the production company name
        company_name = company['name']
        
        # Increment the count for the production company
        if company_name in production_counts:
            production_counts[company_name] += 1
        else:
            production_counts[company_name] = 1

# Create a new DataFrame to store the production company names and their counts
production_df = pd.DataFrame({'Production Company': list(production_counts.keys()), 'Count': list(production_counts.values())})

# Sort the DataFrame by count in descending order
production_df = production_df.sort_values('Count', ascending=False)
print("Production companies by the number of movies produced:")
production_df
```

# Displaying the names of the top 25 production companies by the number of movies produced


```python
# Display the top 25 production companies
top_producers = production_df.head(25)
print("Top 25 production companies by the number of movies produced:")
top_producers
```

# Sorting data by revenue, filtering top 500 movies, and calculating central tendency measures

```python
top_500_revenue = df.sort_values(by='revenue', ascending=False).head(500)
print("Central tendency measures for budget:")
print(top_500_revenue['budget'].describe())
print("Central tendency measures for revenue:")
print(top_500_revenue['revenue'].describe())
print("Central tendency measures for runtime:")
print(top_500_revenue['runtime'].describe())
```

# Outlier analysis using box plots

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 5))
plt.subplot(131)
plt.boxplot(top_500_revenue['budget'])
plt.title("Budget Box Plot")

plt.subplot(132)
plt.boxplot(top_500_revenue['revenue'])
plt.title("Revenue Box Plot")

plt.subplot(133)
plt.boxplot(top_500_revenue['runtime'])
plt.title("Runtime Box Plot")

```

**Interpretations:In the box we can able to see the revenue column has maximum number of outliers and in budget 2 outliers are present.
    

# Displaying movies with above-average runtime from the top 500 movies

```python
average_runtime = top_500_revenue['runtime'].mean()
above_average_runtime = top_500_revenue[top_500_revenue['runtime'] > average_runtime]
print("Movies with above-average runtime:")
print(above_average_runtime[['original_title', 'runtime']])
```
